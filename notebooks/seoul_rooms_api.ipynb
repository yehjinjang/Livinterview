```python
import pandas as pd                         # CSV ì½ê¸°ìš©
from sqlalchemy import create_engine        # DB ì—°ê²°ìš©
from dotenv import load_dotenv              # .env íŒŒì¼ì—ì„œ í™˜ê²½ë³€ìˆ˜ ë¶ˆëŸ¬ì˜¤ê¸°
import os                                   # í™˜ê²½ë³€ìˆ˜ ë¡œë“œì— ì‚¬ìš©

# .env ë¡œë“œ
load_dotenv()

# DB ì—°ê²° ì—”ì§„ ìƒì„±
engine = create_engine(os.getenv("DB_URL"))

print("âœ… DB ì—°ê²° ì„±ê³µ!")

# Seoul_rooms í…Œì´ë¸” ì¡°íšŒ
try:
    query = "SELECT * FROM Seoul_rooms"
    df = pd.read_sql(query, engine)
    
    # ë°ì´í„°í”„ë ˆì„ ì¶œë ¥
    print("ğŸ“Š Seoul_rooms ë°ì´í„°í”„ë ˆì„:")
    display(df.head())
except Exception as e:
    print(f"ğŸš¨ ë°ì´í„° ì¡°íšŒ ì‹¤íŒ¨: {e}")
finally:
    engine.dispose()  # ì—”ì§„ ë¦¬ì†ŒìŠ¤ ì •ë¦¬
```
